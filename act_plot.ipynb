```python
from pathlib import Path
from itertools import product
from matplotlib import pyplot as plt
import torch
from custom_activations import (
    SoftExp,
    BMU,
    ISRU,
    ISRLU,
    PBessel,
    LeakyPReQU,
    Sinusoid,
    Modulo,
    TriWave,
    Gaussian,
)

plt.style.use("default")

out_path = Path("activation_plots")
```

```python
x = torch.arange(-5.0, 5.0, 0.01)
```

```python
vals = (-1.0, -0.5, 0.0, 0.5, 1.0)

for alpha in vals:
    y = SoftExp(alpha)(x).detach()
    plt.plot(x, y, label=f"alpha={alpha}")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.title("SoftExp")
plt.savefig(out_path / "SoftExp.png")
```

```python
y = BMU()(x).detach()
plt.plot(x, y)
plt.title("BMU")
plt.savefig(out_path / "BMU.png")
```

```python
y = ISRU()(x).detach()
plt.plot(x, y)
plt.title("ISRU")
plt.savefig(out_path / "ISRU.png")
```

```python
y = ISRLU()(x).detach()
plt.plot(x, y)
plt.title("ISRLU")
plt.savefig(out_path / "ISRLU.png")
```

```python
y = []
vals = (-1.0, 0.0, 1.0)
for alpha, beta in product(vals, vals):
    y = PBessel(alpha, beta)(x).detach()
    plt.plot(x, y, label=f"alpha={alpha}, beta={beta}")

plt.title("PBessel")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.savefig(out_path / "PBessel.png")
plt.show()  # To display the plot if you are working in an interactive environment.
```

```python
vals = (-1.0, -0.5, 0.0, 0.5, 1.0)

for alpha in vals:
    y = TriWave(alpha)(x).detach()
    plt.plot(x, y, label=f"alpha={alpha}")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.title("TriWave")
plt.savefig(out_path / "TriWave.png")
```

```python
def test(alpha):
    # A parametric triangle wave function
    def f(x):
        return torch.abs(2 * (x / alpha - torch.floor(x / alpha + 0.5)))

    return f


y = []
vals = (-1.0, -0.3, 0.3, 1.0)
for alpha, beta in product(vals, vals):
    y = test(alpha, beta)(x).detach()
    plt.plot(x, y, label=f"alpha={alpha}, beta={beta}")

plt.title("test")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.savefig(out_path / "test.png")
```
